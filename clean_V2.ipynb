```python
import pandas as pd
import dask.dataframe as dd

import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df
```

```python
df = dd.read_csv('listings_ny_2018_09.csv',dtype={'beds':float})
df = df[['id', 'description','host_is_superhost', 'room_type', 'accommodates', 'beds', 'price', \
     'availability_30', 'number_of_reviews', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', \
     'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 'instant_bookable', 'reviews_per_month']].compute()
df['price'] = df['price'].str.replace('[\$\,]', '').astype(float)
df['host_is_superhost'] = df['host_is_superhost'].str.replace('f', '0')
df['host_is_superhost'] = df['host_is_superhost'].str.replace('t', '1')
df['host_is_superhost'] = df['host_is_superhost'].fillna('0')
df['host_is_superhost'] = df['host_is_superhost'].astype(int)
df['instant_bookable'] = df['instant_bookable'].str.replace('f', '0')
df['instant_bookable'] = df['instant_bookable'].str.replace('t', '1')
df['instant_bookable'] = df['instant_bookable'].fillna('0')
df['instant_bookable'] = df['instant_bookable'].astype(int)
df['room_type'] = df['room_type'].str.replace('Private room', '1')
df['room_type'] = df['room_type'].str.replace('Entire home/apt', '1')
df['room_type'] = df['room_type'].str.replace('Shared room', '0')
df = df.fillna(0)
#df.to_csv('listing_ny_2018_09_filtered.csv',index=False)
df.head()
```

```python
df = swap_columns(df,'availability_30','description')
df = swap_columns(df,'reviews_per_month','description')
df.head()
```

```python
df.to_csv('listing_ny_2018_09_filtered_V3.csv',index=False)
```
